```python
OTHER_MOVIE_COLS = ["name", "year", "release_date", "imdb_url"]
GENRE_COLS = [c for c in movies.columns if c not in OTHER_MOVIE_COLS]
```

```python
movies
```

```python
# Total de Géneros en una película
movies[GENRE_COLS].sum(axis=1).value_counts().sort_index()
```

```python
movies[GENRE_COLS].sum(axis=0).sort_values(ascending=False)
```

```python
from scipy.cluster.hierarchy import linkage, leaves_list
from scipy.spatial.distance import pdist

def reorder_corr(corr: pd.DataFrame) -> pd.DataFrame:
    dist = pdist(corr.values)  # pairwise distances between observations in the dataset
    link = linkage(dist, method='complete')  # hierarchical/agglomerative clustering
    idx = leaves_list(link)  # list of leaf node ids
    return corr.iloc[idx, :].T.iloc[idx, :]
```

```python
corr = movies[GENRE_COLS].corr()
corr = reorder_corr(corr)
```

```python
corr.style.background_gradient("RdYlGn", vmin=-1.0, vmax=1.0)
```

```python
from typing import Union
from math import isnan

def cut_imdb(v: Union[str, float]) -> str:
    if isinstance(v, float) and isnan(v):
        return "NAN"
    elif v.startswith("M/title-exact?"):
        return "M"
    elif v.startswith("Title?"):
        return "T"
    elif v.startswith("M/title-exact/"):
        return "MB"
    else:
        return "OTHER"
```

```python
movies.imdb_url.apply(cut_imdb).value_counts()
```
